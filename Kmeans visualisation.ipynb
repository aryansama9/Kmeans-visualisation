```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

change = pd.read_csv("assessment_3_dataset_1.csv")
df = change.rename(columns={"25": "Column 1", "-15": "Column 2", "12": "Column 3"}) #used df (dataframe/dataset) as the name so I don't have to type "dataset" repeatedly

my_cluster = KMeans(n_clusters = 3, n_init = 30) #Starting with 3 clusters and 30 intializations 
df["Cluster"] = my_cluster.fit_predict(df)

make = plt.figure()
scatter_plot = make.add_subplot(111, projection = "3d")
colours = ["orange", "lightgreen", "purple"]
shapes = ["o", "*", "^"] #Circle, star and triangle
#these are just the way its made to generate specific symbols

for cluster in range(3):
    cluster_data = df[df["Cluster"] == cluster]
    scatter_plot.scatter(
        cluster_data["Column 1"], 
        cluster_data["Column 2"], 
        cluster_data["Column 3"],
        c = colours[cluster], 
        marker = shapes[cluster], 
        label = f"Cluster number {cluster + 1}") #Without format and "Cluster" with cluster we won't get the idea of what these points represent
    

scatter_plot.set_title("The three 3d clusters")
scatter_plot.set_xlabel("The first column")
scatter_plot.set_ylabel("The second column")
scatter_plot.set_zlabel("The third column")
scatter_plot.legend()
plt.show()
#From this we can tell the exact centre of the square is where 0 is and not towards the back like how it'd appear on a 2d x and y axis graph (Proving this is a 3 dimensional  x, y z axis)
```
